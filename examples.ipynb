```python
test = Orbit(0.9, 6.0, 0.3, .5, 100)
```

```python
test.geo_coords(2)
```

```python
psi_arr = np.linspace(0, 2 * np.pi, num=5000)
```

```python
res = np.zeros((len(psi_arr), 4))
for i in range(len(psi_arr)):
    psi = psi_arr[i]
    res[i, :] = test.geo_coords(psi)
print(res)
```

```python
t = res[:,0]
r = res[:,1]
theta = res[:,2]
phi = res[:,3]
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
plt.plot(r*np.cos(phi)*np.sin(theta), r*np.sin(phi)*np.sin(theta), r*np.cos(theta))
```
